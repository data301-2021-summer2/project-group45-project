```python
#the mean of the rows
df['Tmean'] = df.iloc[:,2:255].mean(axis = 'columns')
df.head(20)
```

```python
#detmined whether to keep the rows or not
def keep_or_discard(x):
    if (x['Tmean'] > 1):
        return 'keep'
    else:
        return 'discard'

df.apply(keep_or_discard, axis='columns')
```

```python
#apply keep or discard status to the df
df['TStatus'] = df.apply(keep_or_discard, axis='columns')

display(df)
```
